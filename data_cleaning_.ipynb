```python
import pandas as pd
df = pd.read_csv('practiceDataset.xlsx - Sheet1 (1).csv')
df.head()
```

```python
df = df.drop(columns=["Street Number Suffix", "Unit", "Unit Suffix",
                 "Completed Date", "Structural Notification", "Voluntary Soft-Story Retrofit",
                 "Fire Only Permit", "TIDF Compliance", "Site Permit"

])
df.head()
```

```python
df = df.fillna(df.median(numeric_only=True))
df.head()
```

```python
df = df.fillna("Unknown")
df.head()
```

```python
print(df.duplicated())
```

```python
df["Permit Creation Date"] = pd.to_datetime(df["Permit Creation Date"])
df["Filed Date"] = pd.to_datetime(df["Filed Date"])


df.head()
```

```python
df.columns = df.columns.str.strip()
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```
